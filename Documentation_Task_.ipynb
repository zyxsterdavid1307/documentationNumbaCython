# While working with Numpy Arrays


Working Method:
- Write the **Numpy Code** using traditional method

- Then to ***Cythonise*** our code, first load the Cython extension by using the command  `%load_ext cython`
- Then initialise your code  by using magic cell command `%%cython`
- Further initialise all the functions and variables with `cdef` keyword alongwith with their appropriate *datatypes* such as `int`, `str`, `list`, `float`, `double` e.t.c

- For optimisation using ***Numba*** we just need to decoarate each function of our code with  the  `@jit` decorator  also import the modules from numba library using `from numba import jit`
- For ex - 

    ```
    @jit(nopython = True)
    def function_name(func_paramter):
        //--code--//
    ```
Here we can also use several other compilation modes for the `@jit` according to our requirements such as `nogil`, `cache`, `parallel` 
- `@jit` works well with `nopython` mode i.e. without using python interpreter and it's other mode of compilation is `object` mode which uses python interpreter and is slow from the former one

Execution Time:
- To test and compare each one of the compilation techniques use `%timeit`
- For ex - `%timeit func_name(func_parameter)`
- Here in our we need to test the execution time result three times - 

    `%timeit func_name_numpy(func_parameter)`  - pure numpy code

    `%timeit func_name_cython(func_parameter)` - cythonised version

    `%timeit func_name_numba(func_parameter)`  - numba optimised code

Possible Errors:
- While working on Google Colab, execute the calling of the fuction in a new cell (only for cythonised version) - 

 `%timeit func_name_cython(func_parameter)` 

 otherwise it will throw an error - *Expected an identifier got literal*

Conclusion:
- Numba  is  way  much  faster  than the Cython version and the original Numpy code itself
- Cython version of the code is fast in comaparison to the Numpy but not as fast as Numba

- declalring all the functions and variables of our code with `cdef`
- example code -
    ```

    ```
    ```
